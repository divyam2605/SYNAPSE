```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/gdrive/MyDrive/Task2hp/HPSpells.csv")
data
```

First 5 rows


```python
data.head()
```

Non-Hex Rows

```python
data.drop(data[data.Type=="Hex"].index, inplace = True)
data
```

No rows with NaN values in Light column

```python
data.dropna(subset=["Light"], inplace = True)
data
```

Bar Graph

```python
freq = {}
for Type in data.Type.unique():
  c = 0
  for item in data.Type:
    if item == Type:
      c+=1
    freq[Type] = c
  freq  
plt.figure(figsize = (49, 4))
plt.bar(range(len(data.Type.unique())), list(freq.values()))
```
